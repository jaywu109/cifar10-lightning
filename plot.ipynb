```python
import os 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
metrics = pd.read_csv("/workspace/CIFAR10_lightning/cifar10_logs/trial_1/lightning_logs/version_0/metrics.csv")
del metrics["step"]
metrics.set_index("epoch", inplace=True)
metrics.dropna(axis=1, how="all").head()
metrics.drop(columns=["lr-Adam"], inplace=True)
sn.relplot(data=metrics, kind="line")
plt.subplots_adjust(top=0.9)
plt.suptitle('CustomCNN', fontsize = 16)
plt.savefig('CustomCNN.jpg', dpi=500)
```
